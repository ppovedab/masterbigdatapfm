```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import requests
```

```python
complete = pd.read_csv("../data/f18/complete.csv", encoding = 'utf8')
complete.head()
```

```python
interesting_columns = ["ID", "club", "club_logo", "nationality", "flag", "photo"]
df = complete[interesting_columns].copy()
```

```python
df["club_logo_fixed"] = df["club_logo"].str.replace("https://cdn.sofifa.org/", "https://cdn.sofifa.org/1x/")
df["photo_fixed"] = df["photo"].str.replace("https://cdn.sofifa.org/", "https://cdn.sofifa.org/1x/")
```

```python
df.club_logo_fixed = df.club_logo_fixed.fillna("https://upload.wikimedia.org/wikipedia/commons/9/93/Torchlight_help_gray.png")
df.photo_fixed = df.photo_fixed.fillna("https://upload.wikimedia.org/wikipedia/commons/9/93/Torchlight_help_gray.png")
df.club = df.club.fillna("unknown")
```

```python
l = len(df) 
listaDF = range(0, l)

for i in listaDF:
    
    club_logo_url = df.loc[i,"club_logo_fixed"]
    photo_url = df.loc[i,"photo_fixed"]
    flag_url = df.loc[i,"flag"]

    club_logo_img = requests.get(club_logo_url).content
    club_logo_png = "../club/" + df.loc[i,"club"] + ".png"
    with open(club_logo_png, 'wb') as handler:
        handler.write(club_logo_img)
        handler.close()

    photo_img = requests.get(photo_url).content
    photo_png = "../photo/" + str(df.loc[i,"ID"]) + ".png"
    with open(photo_png, 'wb') as handler:
        handler.write(photo_img)
        handler.close()
        
    flag_img = requests.get(flag_url).content
    flag_png = "../flag/" + df.loc[i,"nationality"] + ".png"
    with open(flag_png, 'wb') as handler:
        handler.write(flag_img)
        handler.close()
```
