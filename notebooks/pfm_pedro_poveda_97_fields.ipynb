```python
# Importamos librería Pandas

import pandas as pd
```

```python
# Leemos el dataset

complete = pd.read_csv("../data/f18/complete.csv", encoding = 'utf8')
complete.head()
```

```python
def position(df):
  if (df["prefers_gk"] == True):
    return 'gk'
  elif (df["prefers_cb"] == True) | (df["prefers_lcb"] == True) | (df["prefers_rcb"] == True):
    return 'cb'
  elif (df["prefers_rb"] == True) | (df["prefers_lb"] == True) | (df["prefers_rwb"] == True) | \
  (df["prefers_lwb"] == True):
    return 'wb'
  elif (df["prefers_cdm"] == True) | (df["prefers_rdm"] == True) | (df["prefers_ldm"] == True) | \
  (df["prefers_cm"] == True) | (df["prefers_rcm"] == True) | (df["prefers_lcm"] == True) | \
  (df["prefers_lm"] == True) | (df["prefers_rm"] == True):
    return 'mid'
  elif (df["prefers_lam"] == True) | (df["prefers_ram"] == True) | (df["prefers_cam"] == True):
    return "f_mid"
  elif (df["prefers_lw"] == True) | (df["prefers_rw"] == True):
    return 'wing'
  elif (df["prefers_st"] == True) | (df["prefers_cf"] == True) | (df["prefers_lf"] == True) | \
  (df["prefers_rf"] == True) | (df["prefers_ls"] == True) | (df["prefers_rs"] == True):
    return 'st'
  else:
    return 'oth'

complete['position'] = complete.apply(position, axis=1)
```

```python
# Obtenemos las columnas que nos interesan

columns = ["ID","name","full_name","club","age","league","nationality","eur_value",
           "eur_wage","eur_release_clause","overall","potential","position","preferred_foot", 
           "1_on_1_rush_trait","outside_foot_shot_trait","playmaker_trait","skilled_dribbling_trait",
           "team_player_trait","technical_dribbler_trait","playmaker_speciality","engine_speciality",
           "distance_shooter_speciality","crosser_speciality","strength_speciality","leadership_trait"]
df = complete[columns].copy()
df.head()
```

```python
# Escribimos el resultado en CSV

df.club = df.club.fillna("unknown")
df.league = df.league.fillna("unknown")
df.to_csv('../data/output/fields_2018.csv', sep=',', columns = columns, encoding='utf-8')
```
