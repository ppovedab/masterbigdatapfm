```python
# Importamos librerías Pandas y NumPy

import pandas as pd
import numpy as np
```

```python
# Leemos el dataset

complete = pd.read_csv("../data/f18/complete.csv", encoding = 'utf8')
complete.head()
```

```python
# Nos quedamos con las columnas que nos interesan (de momento, todas) y sustituímos NA por 0s y unknown

interesting_columns = ["ID","name","full_name","club","club_logo","special","age","league","birth_date",
                       "height_cm","weight_kg","body_type","real_face","flag","nationality","photo","eur_value",
                       "eur_wage","eur_release_clause","overall","potential","pac","sho","pas","dri","def","phy",
                       "international_reputation","skill_moves","weak_foot","work_rate_att","work_rate_def","preferred_foot",
                       "crossing","finishing","heading_accuracy","short_passing","volleys","dribbling","curve",
                       "free_kick_accuracy","long_passing","ball_control","acceleration","sprint_speed","agility",
                       "reactions","balance","shot_power","jumping","stamina","strength","long_shots","aggression",
                       "interceptions","positioning","vision","penalties","composure","marking","standing_tackle",
                       "sliding_tackle","gk_diving","gk_handling","gk_kicking","gk_positioning","gk_reflexes",
                       "rs","rw","rf","ram","rcm","rm","rdm","rcb","rb","rwb","st","lw","cf","cam","cm","lm","cdm",
                       "cb","lb","lwb","ls","lf","lam","lcm","ldm","lcb","gk","1_on_1_rush_trait",
                       "acrobatic_clearance_trait","argues_with_officials_trait","avoids_using_weaker_foot_trait",
                       "backs_into_player_trait","bicycle_kicks_trait","cautious_with_crosses_trait",
                       "chip_shot_trait","chipped_penalty_trait","comes_for_crosses_trait","corner_specialist_trait",
                       "diver_trait","dives_into_tackles_trait","diving_header_trait","driven_pass_trait",
                       "early_crosser_trait","fan's_favourite_trait","fancy_flicks_trait","finesse_shot_trait",
                       "flair_trait","flair_passes_trait","gk_flat_kick_trait","gk_long_throw_trait",
                       "gk_up_for_corners_trait","giant_throw_in_trait","inflexible_trait","injury_free_trait",
                       "injury_prone_trait","leadership_trait","long_passer_trait","long_shot_taker_trait",
                       "long_throw_in_trait","one_club_player_trait","outside_foot_shot_trait","playmaker_trait",
                       "power_free_kick_trait","power_header_trait","puncher_trait","rushes_out_of_goal_trait",
                       "saves_with_feet_trait","second_wind_trait","selfish_trait","skilled_dribbling_trait",
                       "stutter_penalty_trait","swerve_pass_trait","takes_finesse_free_kicks_trait","target_forward_trait",
                       "team_player_trait","technical_dribbler_trait","tries_to_beat_defensive_line_trait",
                       "poacher_speciality","speedster_speciality","aerial_threat_speciality","dribbler_speciality",
                       "playmaker_speciality","engine_speciality","distance_shooter_speciality","crosser_speciality",
                       "free_kick_specialist_speciality","tackling_speciality","tactician_speciality","acrobat_speciality",
                       "strength_speciality","clinical_finisher_speciality","prefers_rs","prefers_rw","prefers_rf",
                       "prefers_ram","prefers_rcm","prefers_rm","prefers_rdm","prefers_rcb","prefers_rb","prefers_rwb",
                       "prefers_st","prefers_lw","prefers_cf","prefers_cam","prefers_cm","prefers_lm","prefers_cdm",
                       "prefers_cb","prefers_lb","prefers_lwb","prefers_ls","prefers_lf","prefers_lam","prefers_lcm",
                       "prefers_ldm","prefers_lcb","prefers_gk"]

df = complete[interesting_columns].copy()
df.club = df.club.fillna("unknown")
df = df.fillna(0)
df.head()
```

```python
# Cremos nuevos campos:
#  La posición del jugador para usar como variable target de nuestro modelo predictivo, en formato string y numérico
#  Versión numérica de los campos string work_rate y preferred_foot

def position(df):
  if (df["prefers_gk"] == True):
    return 'gk'
  elif (df["prefers_cb"] == True) | (df["prefers_lcb"] == True) | (df["prefers_rcb"] == True):
    return 'cb'
  elif (df["prefers_rb"] == True) | (df["prefers_lb"] == True) | (df["prefers_rwb"] == True) | \
  (df["prefers_lwb"] == True):
    return 'wb'
  elif (df["prefers_cdm"] == True) | (df["prefers_rdm"] == True) | (df["prefers_ldm"] == True) | \
  (df["prefers_cm"] == True) | (df["prefers_rcm"] == True) | (df["prefers_lcm"] == True) | \
  (df["prefers_lm"] == True) | (df["prefers_rm"] == True):
    return 'mid'
  elif (df["prefers_lam"] == True) | (df["prefers_ram"] == True) | (df["prefers_cam"] == True):
    return "f_mid"
  elif (df["prefers_lw"] == True) | (df["prefers_rw"] == True):
    return 'wing'
  elif (df["prefers_st"] == True) | (df["prefers_cf"] == True) | (df["prefers_lf"] == True) | \
  (df["prefers_rf"] == True) | (df["prefers_ls"] == True) | (df["prefers_rs"] == True):
    return 'st'
  else:
    return 'oth'

def position_int(df):
  if (df["prefers_gk"] == True):
    return 1
  elif (df["prefers_cb"] == True) | (df["prefers_lcb"] == True) | (df["prefers_rcb"] == True):
    return 2
  elif (df["prefers_rb"] == True) | (df["prefers_lb"] == True) | (df["prefers_rwb"] == True) | \
  (df["prefers_lwb"] == True):
    return 3
  elif (df["prefers_cdm"] == True) | (df["prefers_rdm"] == True) | (df["prefers_ldm"] == True) | \
  (df["prefers_cm"] == True) | (df["prefers_rcm"] == True) | (df["prefers_lcm"] == True) | \
  (df["prefers_lm"] == True) | (df["prefers_rm"] == True):
    return 4
  elif (df["prefers_lam"] == True) | (df["prefers_ram"] == True) | (df["prefers_cam"] == True):
    return 5
  elif (df["prefers_lw"] == True) | (df["prefers_rw"] == True):
    return 6
  elif (df["prefers_st"] == True) | (df["prefers_cf"] == True) | (df["prefers_lf"] == True) | \
  (df["prefers_rf"] == True) | (df["prefers_ls"] == True) | (df["prefers_rs"] == True):
    return 7
  else:
    return 99

def work_rate_att_int(df):
  if (df["work_rate_att"] == "Low"):
    return 0
  elif (df["work_rate_att"] == "Medium"):
    return 1
  elif (df["work_rate_att"] == "High"):
    return 2
  else:
    return 99

def work_rate_def_int(df):
  if (df["work_rate_def"] == "Low"):
    return 0
  elif (df["work_rate_def"] == "Medium"):
    return 1
  elif (df["work_rate_def"] == "High"):
    return 2
  else:
    return 99

def preferred_foot_int(df):
  if (df["preferred_foot"] == "Right"):
    return 0
  elif (df["preferred_foot"] == "Left"):
    return 1
  else:
    return 99

df['position'] = df.apply(position, axis=1)
df['position_int'] = df.apply(position_int, axis=1)
df['work_rate_att_int'] = df.apply(work_rate_att_int, axis=1)
df['work_rate_def_int'] = df.apply(work_rate_def_int, axis=1)
df['preferred_foot_int'] = df.apply(preferred_foot_int, axis=1)
```

```python
# Elegimos las variables predictoras (X) y la variable target

x_columns = ["pac","sho","pas","dri","def","phy",
             "skill_moves","weak_foot","work_rate_att_int","work_rate_def_int","preferred_foot_int",
             "crossing","finishing","heading_accuracy","short_passing","volleys","dribbling","curve",
             "free_kick_accuracy","long_passing","ball_control","acceleration","sprint_speed","agility",
             "reactions","balance","shot_power","jumping","stamina","strength","long_shots","aggression",
             "interceptions","positioning","vision","penalties","composure","marking","standing_tackle",
             "sliding_tackle","gk_diving","gk_handling","gk_kicking","gk_positioning","gk_reflexes",
             "rs","rw","rf","ram","rcm","rm","rdm","rcb","rb","rwb","st","lw","cf","cam","cm","lm","cdm",
             "cb","lb","lwb","ls","lf","lam","lcm","ldm","lcb","gk","1_on_1_rush_trait",
             "acrobatic_clearance_trait","argues_with_officials_trait","avoids_using_weaker_foot_trait",
             "backs_into_player_trait","bicycle_kicks_trait","cautious_with_crosses_trait",
             "chip_shot_trait","chipped_penalty_trait","comes_for_crosses_trait","corner_specialist_trait",
             "diver_trait","dives_into_tackles_trait","diving_header_trait","driven_pass_trait",
             "early_crosser_trait","fan's_favourite_trait","fancy_flicks_trait","finesse_shot_trait",
             "flair_trait","flair_passes_trait","gk_flat_kick_trait","gk_long_throw_trait",
             "gk_up_for_corners_trait","giant_throw_in_trait","inflexible_trait","injury_free_trait",
             "injury_prone_trait","leadership_trait","long_passer_trait","long_shot_taker_trait",
             "long_throw_in_trait","one_club_player_trait","outside_foot_shot_trait","playmaker_trait",
             "power_free_kick_trait","power_header_trait","puncher_trait","rushes_out_of_goal_trait",
             "saves_with_feet_trait","second_wind_trait","selfish_trait","skilled_dribbling_trait",
             "stutter_penalty_trait","swerve_pass_trait","takes_finesse_free_kicks_trait","target_forward_trait",
             "team_player_trait","technical_dribbler_trait","tries_to_beat_defensive_line_trait",
             "poacher_speciality","speedster_speciality","aerial_threat_speciality","dribbler_speciality",
             "playmaker_speciality","engine_speciality","distance_shooter_speciality","crosser_speciality",
             "free_kick_specialist_speciality","tackling_speciality","tactician_speciality","acrobat_speciality",
             "strength_speciality","clinical_finisher_speciality","prefers_rs","prefers_rw","prefers_rf",
             "prefers_ram","prefers_rcm","prefers_rm","prefers_rdm","prefers_rcb","prefers_rb","prefers_rwb",
             "prefers_st","prefers_lw","prefers_cf","prefers_cam","prefers_cm","prefers_lm","prefers_cdm",
             "prefers_cb","prefers_lb","prefers_lwb","prefers_ls","prefers_lf","prefers_lam","prefers_lcm",
             "prefers_ldm","prefers_lcb","prefers_gk"]

y_column = ["position_int"]
```

```python
# Creamos los datasets de training y test, con un split de 80 - 20 %

s = int(round(len(df)*0.8))
train = df[:s]
test = df[s+1:]

X_train = train[x_columns] 
X_test = test[x_columns] 
y_train = np.array(train[y_column])
y_test = np.array(test[y_column])
```

```python
# Cargamos las librerías de sklearn para clasificación
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Instanciamos el modelo (k = 10 ,distance como weight y n_jobs -1 para usar toda la potencia de la CPU)
knn = KNeighborsClassifier(10, weights = "distance", n_jobs = -1)

# Ajustamos el modelo
model = knn.fit(X_train, y_train.ravel())

# Aplicamos el modelo en el conjunto de test
pred = model.predict(X_test)

# Evaluamos la precisión del modelo
print accuracy_score(y_test, pred)
```

```python
# Tomamos un ejemplo: Lucas Vázquez
player_index = 211
neighbors = model.kneighbors([df.iloc[player_index][x_columns]],n_neighbors=10)[1]
neighbors = np.reshape(neighbors,(neighbors.shape[1]))
df.iloc[neighbors[0:]][["ID", "name", "overall", "age", "position", "club", "league", "eur_value", "eur_release_clause"]]
```

```python
# El último paso es generar un fichero con todos los futbolistas, sus vecinos y las características
# que queremos mostrar de ellos

# Inicializamos el dataset

l = len(complete) 
n = l * 20
index = range(0,n)
columns = ["ID", "neighbor", "n_ID", "n_name", "n_overall", "n_age", 
           "n_position", "n_club", "n_league", "n_eur_value", "n_eur_release_clause"]
neighborsDF = pd.DataFrame(index=index, columns=columns)

# Recorremos el dataset original con un bucle obteniendo los vecinos de cada futbolista

listaJugadores = range(0, l)
k = 0
n = 21
for i in listaJugadores:
    neighbors = model.kneighbors([df.iloc[i][x_columns]],n_neighbors=n)[1]
    neighbors = np.reshape(neighbors,(neighbors.shape[1]))
    listaNeighbors = range(0, n)
    for j in listaNeighbors:
        neighborsDF.loc[k,"ID"] = df.loc[i,'ID']
        neighborsDF.loc[k,"neighbor"] = j
        neighborsDF.loc[k,"n_ID"] = df.iloc[neighbors[j]]["ID"]
        neighborsDF.loc[k,"n_name"] = df.iloc[neighbors[j]]["name"]
        neighborsDF.loc[k,"n_overall"] = df.iloc[neighbors[j]]["overall"]
        neighborsDF.loc[k,"n_position"] = df.iloc[neighbors[j]]["position"]
        neighborsDF.loc[k,"n_club"] = df.iloc[neighbors[j]]["club"]
        neighborsDF.loc[k,"n_league"] = df.iloc[neighbors[j]]["league"]
        neighborsDF.loc[k,"n_eur_value"] = df.iloc[neighbors[j]]["eur_value"]
        neighborsDF.loc[k,"n_eur_release_clause"] = df.iloc[neighbors[j]]["eur_release_clause"]
        k = k + 1

neighborsDF.head()
```

```python
# Escribimos el resultado en CSV
neighborsDF.to_csv('../data/output/neighbors.csv', sep=',', columns = columns, encoding='utf-8')
```
